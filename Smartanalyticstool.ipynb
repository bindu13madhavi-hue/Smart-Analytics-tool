```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("IPL_Matches_2008_2022.csv")
```

```python
import plotly.express as px
```

```python
column = input("Enter Column Name: ")

px.histogram(
    df,
    x=column,
    title='Histogram'
).show()
```

```python
column = input("Enter Column Name: ")

px.box(
df,
y=column,
title='Box Plot'
).show()
```

```python
x_col = input("X Column: ")
y_col = input("Y Column: ")

px.scatter(
df,
x=x_col,
y=y_col,
title='Scatter Plot'
).show()
```
